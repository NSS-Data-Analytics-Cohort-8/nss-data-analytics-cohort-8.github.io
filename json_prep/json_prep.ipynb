```python
import pandas as pd
import json
```

```python
cohort = pd.read_csv('da8_website_info - Sheet1.csv')
```

```python
cohort.head(2)
```

```python
cohort.columns
```

```python
cohort = cohort.rename(columns = {
    'First': 'firstName',
    'Last': 'lastName',
    'LinkedIn': 'linkedIn',
    'Github': 'github',
    'Resume': 'resume',
    'Email': 'email',
    'Tagline': 'reelThemIn',
    'Bio': 'bio'
})
```

```python
cohort = cohort.sort_values('lastName').reset_index(drop = True)
```

```python
cohort = cohort.reset_index()
```

```python
cohort = cohort.rename(columns = {'index': 'id'})
```

```python
cohort.columns
```

```python
cohort['proImg'] = 'assets/img/placeholder.jpg'
cohort['funImg'] = 'assets/img/placeholder.jpg'
cohort['resume'] = None
```

```python
photo_people = ['Haskins', 'Landers', 'French', 'Ward', 'Nketiah', 'Greer', 'Trask',
               'Haggerty', 'Wesson', 'Nash', 'Shanks', 'Ruiz',
                'Cost', 'Craig', 'Pentecost', 'Ranck', 'Spence', 'Best'
               ]

for person in photo_people:
    cohort.loc[cohort['lastName'] == person, 'proImg'] = \
    'assets/img/' + cohort.loc[cohort['lastName'] == person, 'firstName'].str.lower() + \
    '1.jpg'
    cohort.loc[cohort['lastName'] == person, 'funImg'] = \
    'assets/img/' + cohort.loc[cohort['lastName'] == person, 'firstName'].str.lower() + \
    '2.jpg'
```

```python
resume_people = ['Haskins', 'French', 'Craig', 'Nketiah', 'Spence', 'Best', 'Greer',
                'Trask', 'Haggerty', 'Wesson', 'Nash', 'Shanks', 'Ruiz',
                'Cost', 'Ward']

for person in resume_people:
    cohort.loc[cohort['lastName'] == person, 'resume'] = 'resumes/' + cohort.loc[cohort['lastName'] == person, 'firstName'] + \
    " " + cohort.loc[cohort['lastName'] == person, 'lastName'] + '.pdf'
```

```python
cohort
```

```python
cohort_json = {'cohort': json.loads(cohort.to_json(orient = 'records'))}
```

```python
with open('cohort.json', 'w') as fi:
    json.dump(cohort_json, fi)
```
